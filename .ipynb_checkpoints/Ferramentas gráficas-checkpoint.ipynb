# Matplotlib

Matplotlib representa graficamente seus dados em **Figures** (ou seja, janelas, widgets Jupyter, etc.), cada um dos quais pode conter um ou mais Axes(ou seja, uma área onde os pontos podem ser especificados em termos de coordenadas xy

### A maneira mais simples de criar uma figura com eixos é usando pyplot.subplots

```python
plt.subplots?
```

```python
fig, ax = plt.subplots() # Cria uma figura contendo apenas o eixo
```

#### Passando dados para os eixos

```python
fig, ax = plt.subplots()
ax.plot([-4,-3,-2,-1,0,1,2,3,4], [16,9,4,1,0,1,4,9,16]) #### Passando dados para os eixos
```

#### Na verdade você não precisa criar manualmente a figura e seus eixos, você pode deixar o matplotlib fazer isso

```python
plt.plot([-4,-3,-2,-1,0,1,2,3,4], [16,9,4,1,0,1,4,9,16])
```

## Componentes de uma figura do matplotlib

### Figure

A figura inteira . A figura acompanha toda a criança Axes, um punhado de artistas "especiais" (títulos, lendas das figuras, etc.)

A maneira mais fácil de criar uma nova figura é com o pyplot

#### Uma figura vázia

```python
fig = plt.figure() #Uma figura vázia sem eixos
```

#### Uma figura simples com os eixos

```python
fig, ax = plt.subplots()
```

#### Uma figura com uma grade 2x2 de eixos

```python
fig, axs = plt.subplots(2,2)
axs[0][0].plot([-4,-3,-2,-1,0,1,2,3,4], [16,9,4,1,0,1,4,9,16])
axs[0][1].plot([1,2,3],[2,4,6])
```

### Axes

Isso é o que você pensa como 'um gráfico', é a região da imagem com o espaço de dados. Uma determinada figura pode conter muitos eixos, mas um determinado Axes objeto só pode estar em um Figure.

#### A escala dos eixos podem ser controlados usando:
* axes.set_xlim()
* axes.set_ylim()

#### Os títulos dos eixos podem ser definidos como:
* set_title()
* set_xlabel()
* set_ylabel()

#### Adicionando uma legenda:
* ax.legend()

```python
fig, ax = plt.subplots()

### Fazendo ajustes dos eixos

ax.set_xlim(-4,4) # Setando o limite do eixo x de -4 a 4
ax.set_ylim(0,16) # Setando o limite do eixo y de 1 a 16
ax.set_title('Título Qualquer')
ax.set_xlabel('Grãos de arroz')
ax.set_ylabel('Valor dos grãos')

### Plotando os dados

quadratic = ax.plot([-4,-3,-2,-1,0,1,2,3,4], [16,9,4,1,0,1,4,9,16])
ax.plot([0,1,2,3,4], [0,3,6,9,12])

ax.legend(['Quadrática', 'Linear'], loc='upper left', bbox_to_anchor=(1.05, 1))
```

### Artist

**Basicamente tudo o que você pode ver na figura é um artista** (mesmo as Figure, Axese Axisobjetos). Isso inclui Textobjetos, Line2Dobjetos, collectionsobjetos, Patch objetos ... (você entendeu). Quando a figura é renderizada, todos os artistas são atraídos para a tela . A maioria dos artistas está ligada a um eixo; tal artista não pode ser compartilhado por vários eixos ou movido de um para outro.

## Existem duas maneiras de trabalhar com pyplot
### A interface orientada a objetos e a interface pyplot 

#### Orientada a objetos

Crie explicitamente figuras e eixos e chame métodos neles (o "estilo orientado a objetos (OO)")

```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

#### Usando a interface do pyplot, ou seja, confie no pyplot

Confie no pyplot para criar e gerenciar automaticamente as figuras e eixos e use as funções do pyplot para fazer a plotagem

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

Deve se restringir o pyplot à plotagem interativa (por exemplo, em um caderno Jupyter) e preferir o estilo OO para plotagem não interativa (em funções e scripts que devem ser reutilizados como parte de um projeto maior) .

### Normalmente acontece de você se encontrar fazendo os mesmos gráficos repetidamente mas com conjuntos de dados diferente. O que leva à necessidade de escrever funções especializadas para o gráfico

```python
def my_plotter(ax, eixo_x, eixo_y, parametros_adicionais):
    return ax.plot(eixo_x, eixo_y, **parametros_adicionais)
```

##### Que poderia ser usado como

```python
dados0, dados1, dados2, dados3 = np.random.rand(4,100)
fig, ax = plt.subplots(2,2)
for i in range(0,2):
    for j in range(0,2):
        if (i == 0):
            if (j == 0):
                my_plotter(ax[i][j], dados0, dados1, {'marker': 'x'})
            else:
                my_plotter(ax[i][j], dados0, dados2, {'marker': 'x'})
        else:
            if (j == 0):
                my_plotter(ax[i][j], dados2, dados3, {'marker': 'o'})
            else:
                my_plotter(ax[i][j], dados2, dados1, {'marker': 'o'})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, dados0, dados1, {'marker': 'x'})
my_plotter(ax2, dados0, dados3, {'marker': 'x'})
```
